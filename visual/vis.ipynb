```python
clean_datasets = [13, 14,20, 33, 35, 38, 39, 56, 69, 73, 74, 75, 82, 87, 88,
                  101, 104, 106, 109, 110, 119, 124, 130, 137, 141, 144, 152, 
                  153, 157, 161, 169, 176, 184, 188, 189, 193, 201, 202, 204,
                  206, 207, 210, 211, 212, 214, 218, 244, 246, 253, 256, 273, 276, 297]
```

```python
import pandas as pd
import numpy as np
import plotly.express as px

times = pd.date_range(start="00:00", end="23:30", freq="30min").time


for i in clean_datasets:
# for i in [13]:
    load = torch.load(f'../data/1_processed/{i}.pt', weights_only=False)['load'].reshape(-1,48).mean(dim=0)
    pv = torch.load(f'../data/1_processed/{i}.pt', weights_only=False)['pv'].reshape(-1,48).mean(dim=0)
    df = pd.DataFrame({'Times': times, "Load": load, "PV": pv})

    fig = px.line(df,x='Times', y=['Load', 'PV'], title=f'Customer {i}')
    fig.update_layout(
        xaxis_title="Time [h]",
        yaxis_title="Power [kW]",
        template="plotly_white",
        width=600,
        margin=dict(l=10, r=10, t=30, b=10),
        xaxis=dict(dtick=6),
        legend_title_text=""
    )
    fig.show()
    # fig.write_image('load_pv.pdf')
```
